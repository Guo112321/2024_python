```python
df = pd.read_csv("BostonHousing.csv")
df.head()
```

****In CRIM column, there are 5 data points which are detected as outliers.****

```python
high = round(df['crim'].mean()+3*df['crim'].std(),3)
low = round(df['crim'].mean()-3*df['crim'].std(),3)

print('Highest value of crim',high)
print('Lowest value of crim', low)
df[(df['crim']>high) | (df['crim']<low)]
```
